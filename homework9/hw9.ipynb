# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Your code here

def plot_functions():
    # Domain: Set a range of x values from -10 to 10
    x_values = np.linspace(-10, 10, 400)
    
    # Function 1: y = x^2
    y_values_1 = x_values ** 2
    
    # Function 2: y = x^0.5
    y_values_2 = np.sqrt(x_values)
    
    # Create the plot with a figure size of 10x8
    plt.figure(figsize=(10, 8))
    
    # Plot first function with red solid line
    plt.plot(x_values, y_values_1, label=r'$y = x^2$', color='red', linestyle='-', linewidth=2)
    
    # Plot second function with blue dashed line
    plt.plot(x_values, y_values_2, label=r'$y = x^0.5$', color='blue', linestyle='--', linewidth=2)
    
    # Label the x and y axiss
    plt.xlabel('x values', fontsize=14)
    plt.ylabel('y values', fontsize=14)
    
    # Add title
    plt.title('Plot of $y = x^2$ and $y = x^0.5$', fontsize=16)
    
    # Add legend
    plt.legend(fontsize=12)
    
    # Display plot
    plt.show()
    
    # Return x values and corresponding y values as lists
    return x_values.tolist(), y_values_1.tolist(), y_values_2.tolist()

# Call the function and get the values
x_vals, y_vals_1, y_vals_2 = plot_functions()
```

Call your **first** function in the cell below.

```python
# Your code here

# Function 1: Plotting two subplots side by side (horizontal)
def plot_horizontal(x_range):
    """
    This function generates two subplots side by side (horizontal layout) for the functions:
    h(x) = cos(x) on the left, and k(x) = sin(x) on the right.
    
    Parameters:
        x_range (tuple): A tuple specifying the range of x values (start, end).
    
    Returns:
        None
    """
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], 400)
    
    # Function 1: h(x) = cos(x)
    h_x = np.cos(x)

    # Function 2: k(x) = sin(x)
    k_x = np.sin(x)

    # Create a figure with two subplots horizontally
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns
    
    # Plot for the first subplot (h(x) = cos(x))
    axs[0].plot(x, h_x, label=r'$h(x) = \cos(x)$', color='blue')
    axs[0].set_title(r'Plot of $h(x) = \cos(x)$')
    axs[0].set_xlabel('x values')
    axs[0].set_ylabel('h(x) values')
    axs[0].legend()

    # Plot for the second subplot (k(x) = sin(x))
    axs[1].plot(x, k_x, label=r'$k(x) = \sin(x)$', color='red')
    axs[1].set_title(r'Plot of $k(x) = \sin(x)$')
    axs[1].set_xlabel('x values')
    axs[1].set_ylabel('k(x) values')
    axs[1].legend()

    # Display the plots
    plt.tight_layout()
    plt.show()
```

Call your **second** function in the cell below.

```python
# Your code here

# Function 2: Plotting two subplots stacked on top of each other (vertical)
def plot_vertical(x_range):
    """
    This function generates two subplots stacked vertically (vertical layout) for the functions:
    h(x) = cos(x) on the top, and k(x) = sin(x) on the bottom.
    
    Parameters:
        x_range (tuple): A tuple specifying the range of x values (start, end).
    
    Returns:
        None
    """
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], 400)
    
    # Function 1: h(x) = cos(x)
    h_x = np.cos(x)

    # Function 2: k(x) = sin(x)
    k_x = np.sin(x)

    # Create a figure with two subplots vertically
    fig, axs = plt.subplots(2, 1, figsize=(8, 12))  # 2 rows, 1 column
    
    # Plot for the first subplot (h(x) = cos(x))
    axs[0].plot(x, h_x, label=r'$h(x) = \cos(x)$', color='blue')
    axs[0].set_title(r'Plot of $h(x) = \cos(x)$')
    axs[0].set_xlabel('x values')
    axs[0].set_ylabel('h(x) values')
    axs[0].legend()

    # Plot for the second subplot (k(x) = sin(x))
    axs[1].plot(x, k_x, label=r'$k(x) = \sin(x)$', color='red')
    axs[1].set_title(r'Plot of $k(x) = \sin(x)$')
    axs[1].set_xlabel('x values')
    axs[1].set_ylabel('k(x) values')
    axs[1].legend()

    # Display the plots
    plt.tight_layout()
    plt.show()
```

```python
# Import your module and/or your code here

def plot_image_with_colormaps():
    """
    This function generates a random 2D image (using Gaussian noise) and displays
    the image using several different colormaps. The colormaps include 'viridis', 
    'plasma', 'inferno', 'magma', and 'cividis'.
    
    Each subplot includes a title, axis labels, and a colorbar (legend).
    
    No inputs and no outputs; it displays the plot directly.
    """
    # Generate a random image using a Gaussian distribution
    image_data = np.random.randn(100, 100)  # 100x100 image of random values
    
    # List of colormaps to use for the image
    colormaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
    
    # Create a figure to hold multiple subplots (one for each colormap)
    fig, axs = plt.subplots(1, len(colormaps), figsize=(16, 5))
    
    # Overall title for the figure
    fig.suptitle('Different Colormaps', fontsize=16)

    # Loop through the colormaps and plot the image with each colormap
    for i, cmap in enumerate(colormaps):
        # Display the image with the current colormap
        im = axs[i].imshow(image_data, cmap=cmap)
        
        # Add a title to each subplot
        axs[i].set_title(f'Colormap: {cmap}')
        
        # Add axis labels
        axs[i].set_xlabel('X-axis')
        axs[i].set_ylabel('Y-axis')
        
        # Add a colorbar (as a legend) to each subplot
        fig.colorbar(im, ax=axs[i], fraction=0.046, pad=0.04)  # Add colorbar to each subplot
        
        # Optional: Add a dummy legend to the subplot (if needed)
        axs[i].legend([cmap], loc='upper right', fontsize=8)  # Adding colormap name as a "legend"
        
        # Turn off axis lines for a cleaner look
        axs[i].axis('off')  

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)  # Adjust the top to make space for the main title
    plt.show()
```
